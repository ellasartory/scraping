```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Provide the correct file path to your Excel file
file_path = r"C:\Users\misha\jupyter\year 2 - data set linkage\Outscraper FULL VERSION.xlsx"

# Load the Excel file into a pandas DataFrame
data = pd.read_excel(file_path)
df = data

# Describe the dataset
summary_statistics = data.describe()
print(summary_statistics)

# Histplot for average ratings of mobility hubs
sns.histplot(data = df, x = "rating", kde=True)
plt.show()

# Histplot for ratings to the mobility hubs by users
sns.histplot(data = df, x = "review_rating", kde=True)
plt.show()

# A scatterplot, which isnt really that usefull for this data-set.
sns.scatterplot(data = df, x='rating', y='reviews')
plt.show()

# A box plot for the review_ratings from 1-5 and the amount of reviews of that number given)
sns.boxplot( data=df, x='review_rating', y='reviews')
plt.show()





```
