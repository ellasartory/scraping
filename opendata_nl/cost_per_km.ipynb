```python
import pandas as pd
import numpy as np
```

```python

GDP_df = pd.read_csv('cbs_GDP.csv', sep=';')
GDP_df = GDP_df.drop(columns='Periods')
GDP_df = GDP_df.transpose()

# renaming the columns by creating a new dataframe
headers = GDP_df.iloc[0]
new_GDP = pd.DataFrame(GDP_df.values[1:], columns=headers) 

# renaming the provinces to match the other data set
new_GDP = new_GDP.rename(columns={
    "The Netherlands": "the Netherlands",
    "Frysl√¢n (PV)": "Friesland (PV)",
    "Noord-Holland (PV)": "North Holland (PV)",
    "Zuid-Holland (PV)": "South Holland (PV)",
    "Noord-Brabant (PV)": "North Brabant (PV)"
})

# Set the 'Regions' column
new_GDP.insert(0, 'Regions', ["GDP (market prices) (million euros)", "GDP per capita (euros)"], True)


# Create private and shared cost lists
private = ["Car = 0.39", "Bike = 0.0373"]
shared = ["Car = 0.3", "Bike = 0.1176"]

# Create new columns and assign values
new_GDP["Cost per km for privately owned vehicle in ct."] = private
new_GDP["Cost per km for shared vehicle (Hely) in ct."] = shared

# Rearrange the columns
cols = ['Cost per km for privately owned vehicle in ct.', 'Cost per km for shared vehicle (Hely) in ct.', 'Regions'] + \
       [col for col in new_GDP if col != 'Regions' and col not in ['Cost per km for privately owned vehicle in ct.', 'Cost per km for shared vehicle (Hely) in ct.']]

new_GDP = new_GDP[cols]

display(new_GDP)

```
