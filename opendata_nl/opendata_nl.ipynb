```python
%pip install requests
%pip install --user cbsodata 
import pandas as pd
import numpy as np
import statistics
import requests
import cbsodata
```

```python
# Downloading table list
toc = pd.DataFrame(cbsodata.get_table_list())

# Downloading entire dataset 
data = pd.DataFrame(cbsodata.get_data("84711ENG"))

#cleaning and limiting data to relevant parts 
conditions = (
    (data['Population'] == "Population 6 years or older") &
    (data['RegionCharacteristics'] == "The Netherlands") &
    (data['Margins'] == "Value") &
    (data['Periods'] == "2022") &
    (data['ModesOfTravel'] != "Walking") &
    (data['ModesOfTravel'] != "Other")
)

# Apply the filtering conditions
filtered_data = data[conditions]

# Display the filtered data
print(filtered_data)

# Downloading metadata
metadata = pd.DataFrame(cbsodata.get_meta("84711ENG", "DataProperties"))
print(metadata[['Key','Title']])

filtered_data.to_csv("trip_data.csv", index=False)

# Save metadata to a CSV file
metadata.to_csv("metatripdata.csv", index=False)
```

```python
# Replace 'cbs_GDP.csv' with the actual path to your CSV file
GDP_file_path = 'cbs_GDP.csv'

# Read the CSV file into a DataFrame
GDP = pd.read_csv(GDP_file_path)

print(GDP)

# Convert the 'million euros' column to numeric
GDP['GDP per capita (euros)'] = pd.to_numeric(GDP['GDP per capita (euros)'], errors='coerce')

# Create a DataFrame with numeric values
numeric_df = GDP.apply(pd.to_numeric, errors='coerce')

# Calculate the variance
variance_1 = np.var(numeric_df)

print("Variance of the merged data from Opendata.nl:", variance_1)

# Calculate the mean
mean_values = numeric_df.mean()
print(f"Mean: {mean_values}")

# Calculate the standard deviation
std_dev = np.sqrt(variance_1)
print(f"This is the standard deviation: {std_dev}")

```
