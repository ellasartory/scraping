```python
%pip install requests
%pip install --user cbsodata 
import pandas as pd
import numpy as np
import statistics
import requests
import cbsodata
```

```python
# Downloading table list
toc = pd.DataFrame(cbsodata.get_table_list())

# Downloading entire dataset (can take up to 30s)
data = pd.DataFrame(cbsodata.get_data("84711ENG"))
# filtered_data = data[(data['Population'] == "Population 6 years or older") & (data['RegionCharacteristics'] == "The Netherlands") & (data['Margins'] == "Value") & (data['Periods'] == "2022")]
# filtered_data = filtered_data[(filtered_data.ModesOfTravel != "Walking") & filtered_data(filtered_data.ModesOfTravel != "Other")]

conditions = (
    (data['Population'] == "Population 6 years or older") &
    (data['RegionCharacteristics'] == "The Netherlands") &
    (data['Margins'] == "Value") &
    (data['Periods'] == "2022") &
    (data['ModesOfTravel'] != "Walking") &
    (data['ModesOfTravel'] != "Other")
)

# Apply the filtering conditions
filtered_data = data[conditions]

# Display the filtered data
print(filtered_data)

# Downloading metadata
metadata = pd.DataFrame(cbsodata.get_meta("84711ENG", "DataProperties"))
print(metadata[['Key','Title']])

filtered_data.to_csv("trip_data.csv", index=False)

# Save metadata to a CSV file
metadata.to_csv("metatripdata.csv", index=False)
```
